```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="Peter Heslin", timeout=None)
```

```python
location = geolocator.geocode("New Elvet, Durham")
location
```

```python
location = geolocator.geocode("Durham")
location
```

```python
locations = geolocator.geocode("Durham", exactly_one=False)
locations
```

```python
location = geolocator.geocode("Durham, USA")
location
```

```python
location.latitude, location.longitude
```

```python
durhams = []
locations = geolocator.geocode("Durham", exactly_one=False)
for location in locations:
    print(location)
    durhams.append([location.latitude, location.longitude])
```

```python
durhams
```

## Folium

```python
import folium
```

```python
chem = geolocator.geocode("Chemistry Cafe, Durham, England")
chem
```

```python
durham_map = folium.Map(location=[chem.latitude, chem.longitude], zoom_start=18)
durham_map
```

```python
folium.Marker(location=[chem.latitude, chem.longitude], popup="Our classroom").add_to(durham_map)
durham_map
```

```python
loc = geolocator.geocode("United States")
```

```python
usa_map = folium.Map(location=[loc.latitude, loc.longitude], zoom_start=4)
usa_map
```

```python
for loc in durhams:
    folium.Marker(location=[loc[0], loc[1]], popup="Durham").add_to(usa_map)
usa_map
```

```python
with open('dead_places.txt') as f:
    dead_places = f.read().splitlines()
```

```python
dead_places
```

```python
loc = geolocator.geocode("Dublin, Ireland")
```

```python
dub_map = folium.Map(location=[loc.latitude, loc.longitude], zoom_start=8)
dub_map
```

```python
locs = []
for place in dead_places:
    location = geolocator.geocode(place, exactly_one=True)
    print(place, location)
    if location:
        locs.append([location.latitude, location.longitude, place])
```

```python
for loc in locs:
    folium.Marker(location=[loc[0], loc[1]], popup=loc[2]).add_to(dub_map)
dub_map
```

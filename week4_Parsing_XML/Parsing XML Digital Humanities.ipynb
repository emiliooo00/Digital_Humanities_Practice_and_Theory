# Parsing XML
## MDS/DH

We saw how text is stored in a file on a computer: 
* At the lower level, Unicode describes how each character is mapped to a sequence of 1s and 0s.

* At a higher level, certain characters are used to signify markup, which is metadata that describes the structure of the text, which has implications for how the text is to be displayed.

* On the level of encoding, how the text is stored in a file is the same as how it is represented in the computer's memory.

* On the level of markup, however, the way the computer represents the structure of the document is completely different to how it is written to disk (serialized).

* After the computer has parsed the file and put it in memory, there are no more pointy brackets.  Instead there is a data structure which looks like a tree.

## XML as a tree

* When you parse an XML file (or a HTML file), it is transformed in memory into a tree of objects defined by the DOM (Document Object Model)  specification.

* This provides a uniform way to pull information out or change the structure, across computer languages (e.g. Javascript).

* Because the elements of valid XML are nested, each element contains a mixture of text and other elements, which may themselves contain a text and other elements, ad infinitum.

* Each element is a child that sits inside one parent, up to the root element.

* Each part of the tree is a node (element, text or attribute)

* Note that an element can be empty, like this: `<lineBreak></lineBreak>`, which can be abbreviated like this: `<lineBreak/>`.

## For Example
```html
<html xmlns="http://www.w3.org/1999/xhtml" >
  <head>
    <title>My title</title>
  </head>
  <body>
    <h1>A heading</h1>
    <a href="http://www.durham.ac.uk">Link text</a>
   </body>
</html>
```

## Standard DOM methods

Each thing in the tree is an object.  All of these DOM objects have methods that you can call to find out about them and to modify the tree in memory.  

These methods are standardized across computer languages; only the object syntax differs.

* Determine the name and type of a given node.

* List all the children of a given node.

* Given a node, get its next sibling or parent.

* Move and delete nodes.

* Create a new element or text node; insert it into the tree.

* Iterate over all of the elements with a given name.

* Find an element with a given id attribute.

* Iterate over all of the text node descendants of an element.

## XPATH

* A language for matching particular nodes: a path down the tree from its root

* An exact path: /html/head/title

* An element anywhere: //h1

* Nesting: //div/p or //div//p or //div/*

* Attributes: //*[@name="My Name"]

* Text nodes: //p/text()

* A useful shorthand in combination with DOM traversal methods.

## Namespaces

* All XML element names can belong to a namespace, which is a defined set of elements with their meanings.

* This is defined by a URL, where you can look up the meaning of that element.

* The root element often has an xmlns attribute which declares the default namespace for the document.

* All elements in that document have official names prefixed by that namespace URL.

* When you are looking for elements, remember this stipulation.

* Individual elements can non-default namespace URL, which indicates that this element comes from a different set. 

## Text Encoding Initiative (TEI)

* An XML schema against which documents can be validated.

* Uniformity of tagging and structure.

* Flexible, extensive, stable and ubiquitous.

* Well documented.

* But  There are multiple ways of doing the same thing.

* Overlapping markup in XML requires painful workarounds. 

* Permits/encourages excessive/subjective markup.
